```python
import numpy as np
import matplotlib.pyplot as plt

# Define differential equation
def f(x, y):
    return x - y

# Define initial conditions
x0 = 0
y0 = 1

# Define step size and number of steps
h = 0.1
n = 100

# Solve differential equation using Euler method
x = np.zeros(n+1)
y = np.zeros(n+1)
x[0] = x0
y[0] = y0
for i in range(n):
    x[i+1] = x[i] + h
    y[i+1] = y[i] + h*f(x[i], y[i])

# Plot results
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
